```python
import re
import nltk
from nltk.tokenize import  word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

nltk.download('stopwords')
nltk.download('punkt')

stopwords_ru = stopwords.words("russian")

def to_lowercase(data):
    #Приводим все слова к нижнему регистру
    data = data.lower()
    return data

def noise_remove(data, remove_numbers=True):
    # Удалим ссылки
    data = re.sub(r"(\w+:\/\/\S+)", " ", data)

    # Удалим все символы не являющиеся буквенными символами, числами и пробелами/табами/переносми строки
    data = re.sub(r"([^0-9A-Za-zА-Яа-я])", " ", data)

    # Удалим числа, если необходимо
    if remove_numbers:
        data = re.sub(r"\d+", " ", data)
    return data

def tokenize(text):
    words = text.split()
    for elem in words:
        if len(elem) < 3:
            words.remove(elem)
    return words
```

```python
data['lowered_description'] = data['description'].apply(lambda x: to_lowercase(str(x)))
data['no_noise_description'] = data['lowered_description'].apply(lambda x: noise_remove(x))
data['tokenize_description'] = data['no_noise_description'].apply(lambda x: tokenize(x))
```

```python
def sent_vec(sent):
    vector_size = (300,)
    wv_res = np.zeros(vector_size)
    # print(wv_res)
    ctr = 1
    for w in sent:
        if w in navec:
            ctr += 1
            wv_res += navec[w]
    wv_res = wv_res/ctr
    return wv_res
```

```python
data['ebmendings'] = data['tokenize_description'].apply(lambda x: sent_vec(x))
```

```python
emb = np.array(data['ebmendings'].tolist())
emb.shape
```

```python
from sklearn.preprocessing import OneHotEncoder


ohe = OneHotEncoder(sparse=False, dtype='int')
ohe.fit(data['label'].values.reshape(-1, 1));
Y_toy_ohe = ohe.transform(data['label'].values.reshape(-1, 1))


Y_toy_ohe
```

```python
import keras
```

```python
import tensorflow as tf
```

```python
photo_input = keras.Input(shape=(200 , 200, 3), name='photo')  # Последовательность целых чисел переменной длины
text_input = keras.Input(shape=(300,), name='text')  # Последовательность целых чисел переменной длины

# Сокращаем последовательность вложенных слов заголовка до одного 128-мерного вектора
text_features = keras.layers.Dense(200,  activation='relu')(text_input)
# Сокращаем последовательность вложенных слов заголовка до одного 32-мерного вектора
body_features = keras.layers.Conv2D(20, 3, activation = 'sigmoid')(photo_input)
bn = keras.layers.BatchNormalization()(body_features)
body_features2 = keras.layers.MaxPooling2D(3)(bn)
body_features3 = keras.layers.Conv2D(4, 3, activation = 'sigmoid')(body_features2)
body_features4 = keras.layers.MaxPooling2D(3)(body_features3)

body_features7 = keras.layers.Flatten()(body_features4)
body_features6 = keras.layers.Dense(100,  activation='sigmoid')(body_features7) 

text_features2 = keras.layers.Dense(100,  activation='sigmoid')(text_features) 
# Объединим все признаки в один вектор с помощью конкатенации
x = keras.layers.concatenate([text_features2, body_features6])

# Добавим классификатор отделов прогнозирующий на признаках
department_pred = keras.layers.Dense((23), activation='softmax', name='department')(x)

# Создание сквозной модели, предсказывающей приоритет и отдел
model = keras.Model(
    inputs=[photo_input, text_input],
    outputs=[department_pred]
)

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss=['categorical_crossentropy'],
)
```
